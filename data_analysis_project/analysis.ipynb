```python
import pandas as pd
from pandasql import sqldf
```

```python
# Load CSV into pandas
df = pd.read_csv("Net_User_data(in).csv")
```

```python
# Define SQL execution function# Example SQL query
# Example SQL query

# Define SQL execution function
pysqldf = lambda q: sqldf(q, globals())
query = "SELECT * FROM df;"
result = pysqldf(query)

print(result)

```

```python
#calcuates the average usage in
# Convert 1990 column to numeric (force numbers, ignore errors)
df["1990"] = pd.to_numeric(df["1990"], errors='coerce')
query= "SELECT AVG(1990) AS avg_usage_1990 FROM df"


avg_usage = pysqldf(query)
print(avg_usage)
```
